```python
import pandas as pd
import csv
data = pd.read_csv('titanic//train.csv')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
data = pd.read_csv('titanic//train.csv')
df = pd.DataFrame(data)

# Gender Class
df.loc[df['Sex'] == "male", 'Sex'] = 0
df.loc[df['Sex'] == "female", 'Sex'] = 1

# Embarked Class
df.loc[df['Embarked'] == "C", 'Embarked'] = 0
df.loc[df['Embarked'] == "Q", 'Embarked'] = 1
df.loc[df['Embarked'] == "S", 'Embarked'] = 2
df["Embarked"]=df["Embarked"].fillna(2)

# Fare Class

# Replace empty row with median fare

from numpy import median
median_fare = median(df["Fare"])

df["Fare"]=df["Fare"].fillna(median_fare)

# Divide into three types of payments
# 0 for less than 10
# 1 for more than 10 but less than or equal to 40
# 2 for over 40 
df.loc[df['Fare']<=10, 'Fare'] = 0
df.loc[(df['Fare'] > 10) & (df['Fare'] <= 40), 'Fare'] = 1
df.loc[df['Fare'] > 40, 'Fare'] = 2

df['Fare'] = df['Fare'].astype(int)


#age

df["Age"]=df["Age"].fillna(0)
median_age= median (df["Age"])
print(median_age)
df["Age"]=df["Age"].replace(0, median_age)

print(df)
```

```python
X = data.drop([ "Cabin", "Name", "Ticket", "Survived"], axis=1)
y = data["Survived"]

print(X)
print(y)

# split into train and test set
X_train, X_test, y_train, y_test = train_test_split(
    X, y, stratify=y, test_size=0.10, random_state=100
)
```

```python
# create the classifier
classifier = RandomForestClassifier(n_estimators=100)

# Train the model using the training sets
classifier.fit(X_train, y_train)

```

```python
# predictin on the test set
y_pred = classifier.predict(X_test)
```

```python
print(y_pred)
print(y_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
```
